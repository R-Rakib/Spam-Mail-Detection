```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/spam.csv',encoding='latin-1')
df
```

```python
df=df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
df.head()
```

```python
df=df.rename(columns={'v1':'Class','v2':'Message'})
df.head()
```

```python
df['Class']=df['Class'].map({'ham':0, 'spam':1})
df['Class']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
x=df['Message']
x
```

```python
y=df['Class']
y
```

```python
df['Length']=df['Message'].apply(len)
```

```python
df.head()
```

```python
import plotly.express as px
fig = px.histogram(df, x='Length',color='Class', marginal='rug',
                   title='Message Length')
fig.update_layout(
    xaxis_title='Length',
    yaxis_title='Frequency',
    showlegend=True)
```

```python
x=cv.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model=MultinomialNB()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
text='07732584351 - Rodger Burns - MSG = We tried to call you re your reply to our sms for a free nokia mobile + free camcorder. Please call now 08000930705 for delivery tomorrow'
data=[text]
vect=cv.transform(data).toarray()
result=model.predict(vect)
if result==1:
 print('This a spam email')
if result ==0:
 print('This is not spam mail')
```
